## Tupla

### Tupla é uma estrutura imutavel, ou seja os dados dessa estrutura não são alterados, não pode mudar, segue o exemplo em código abaixo

```python
tupla = ('string')

print(tupla)
```

### Podemos converter em sequencia ou interagir usando a função *tuple*

```python
tup = tuple(tupla)
tup
```

### É possivel acessar o elemento em uma tupla, igual o exemplo em baixo (obseverção o index em python começa com [0])

```python
tup[0]
```

### Ao definir tuplas dentro de expressões mais complicadas, geralmente é necessário colocar os valores entre parênteses, como neste exemplo de criação de uma tupla de tuplas

```python
tupla_dentro_tupla = ((4,5,6),(7,8))

print(tupla_dentro_tupla[0])
print(tupla_dentro_tupla[1])
```

### Embora os objetos armazenados em uma tupla possam ser mutáveis, uma vez criada a tupla não é possível modificar qual objeto está armazenado em cada slot

```python
tup = tuple(['foo', [1, 2], True])

tup[2] = False
```
